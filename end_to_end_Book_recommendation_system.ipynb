## *Recommendation System Types:*
### 1. Popularity based recomendation:
This system recommends a product/content based on the popularity of that product/content on a particular platform. eg. Top 25 IMDB rated movies, Youtube trending page, etc. It does not consider preferences of user.

### 2. Content/Product based recommendation:
This system recommends content/product based on the previous likes of the user. It considers the preferences of the user.

### 3. Collaborative filtering based recommendation:
This system recommends content/product based on the user ratings. In simple words, consider plotting a graph of ratings for a particular product/content, we find distance (Euclidean, etc) between other product/content in the same plane, the least distance one is recommended.
For more info visit: https://developers.google.com/machine-learning/recommendation/collaborative/basics

### 4. Hybrid based recommendation:
This system considers above all systems and make one recommendation out of them with giving certain weightages.

## Let's make a `Popularity-Based Recommendation`

## Let's make a `Collaborating-filtering based Recommendation`

```python
ratings_name_df
```

```python
ratings_name_df.groupby("User-ID").count()["Book-Rating"]
```

```python
# Let's consider users with more than 200 ratings on a book
ratings_name_df.groupby("User-ID").count()["Book-Rating"] > 200
```

```python
x = ratings_name_df.groupby("User-ID").count()["Book-Rating"] > 200
users_200plus_ratings = x[x].index
```

```python
filtered_ratings = ratings_name_df[ratings_name_df["User-ID"].isin(users_200plus_ratings)]
filtered_ratings.shape
```

```python
filtered_ratings.head(10)
```

```python
y = filtered_ratings.groupby("Book-Title").count()["Book-Rating"] >= 50
famous_books = y[y].index
```

```python
final_ratings = filtered_ratings[filtered_ratings["Book-Title"].isin(famous_books)]
final_ratings
```

```python
final_ratings.drop_duplicates()
```

```python
pivot = final_ratings.pivot_table(index="Book-Title", columns="User-ID", values="Book-Rating")
pivot.fillna(0, inplace=True)
pivot
```

```python
# We will find distance between each rows in vector form with the help of cosine-similarity
from sklearn.metrics.pairwise import cosine_similarity
similarity_score = cosine_similarity(pivot)
cosine_similarity(pivot).shape    # This gives Euclidean distance of 706 books with 706 books.
```

*NOTE:* To know about `cosine-similarity` click:https://en.wikipedia.org/wiki/Cosine_similarity

```python
similarity_score
```

```python
# Let's make a recommendation-function
import numpy as np

def recommend(book_name, num_of_recommendations=5):
  """This function takes book_name as an input and gives suggesions as an output based of our saved pivot table.
  """
  index = np.where(pivot.index==book_name)[0][0]    # Getting index of the book name from pivot dataframe.
  similar_books = sorted(list(enumerate(similarity_score[index])), key=lambda x: x[1], reverse=True)[1:num_of_recommendations+1]
  
  recommendations = []

  for i in similar_books:
    item = []
    temp = books_df[books_df["Book-Title"] == pivot.index[i[0]]]
    item.extend(list(temp.drop_duplicates('Book-Title')['Book-Title'].values))
    item.extend(list(temp.drop_duplicates('Book-Title')['Book-Author'].values))
    item.extend(list(temp.drop_duplicates('Book-Title')['Image-URL-M'].values))
    recommendations.append(item)

  return recommendations
```

```python
recommend("Message in a Bottle", num_of_recommendations=1)
```
